## Centrality Measures

### Degree centrality

The Degree centrality indicates the number of links incident upon the specific node. Here, they indicate the most populous intersections in the node. This transaltes to that specific node having higher interactions.

In other words, nodes with higher degree centralities are more likely to be center of attention or more likely to have a VFOA (Visual Focus of attention) within them or at least on the path to reach them.

```python
from networkx.algorithms.centrality import degree_centrality
community=GNcom
G1=G
```

```python
for i in range(len(community)):
    g=G1.subgraph(community[i])
    deg=degree_centrality(g)
    k=max(deg,key=deg.get)
    n=[]
    for key,val in deg.items():
        if(val==deg.get(k)):
            n.append(key)
    print("Community "+str(i+1)+": "+str(n))
```

Thus, we have found the intersections with the most number or roads leading to or from them.

### Betweenness centrality
The Betweenness centrality here gives us the most travelled through intersections. In other words, these nodes are the most travelled through.

```python
from networkx.algorithms.centrality import betweenness_centrality
community=GNcom
G1=G
```

```python
for i in range(len(community)):
    g=G1.subgraph(community[i])
    bet=betweenness_centrality(g)
    k=max(bet,key=bet.get)
    n=[]
    for key,val in bet.items():
        if(val==bet.get(k)):
            n.append(key)
    print("Community "+str(i+1)+": "+str(n))
```

We have found the intersections with the highest number of traffic flowing through them in each comunity. For instance, intersection '70' in community 1 is the busiest intersection.

But these centrality measures do not give the whole idea. This dataset is unweighted. Thus we assume that the traffic density is uniform. This is very rarely the case. But, there are other centrality measures and datasets we can use to demonstrate the locations (or people or object) that can act as Centre of Attention.

## Why this dataset?

In the roadNet dataset, the sole drawback was the lack of weights to represent the flow of traffic. So larger the weight, higher the traffic through that road. The exact same principle applies here. The weights are directly proportional to the amount of time two people spend interacting.

So, when we find centrality measures taking into account these weights, we would getter better results than we did with roadNet. The nodes here represent an employee. Thus, eigen-vector centrality of nodes here would give us an idea as to who is the most influential person in the network or each community. This person would be the focus of attention for us. 

```python
eig=nx.eigenvector_centrality(G1,max_iter=500,weight='weight')
```

```python
max_key=max(eig,key=eig.get)
n=[]
for k,v in eig.items():
    if v==eig[max_key]:
        n.append(k)
```

# Fairness-Goodness

```python
def initiliaze_scores(G):
    fairness = {}
    goodness = {}
    
    nodes = G.nodes()
    for node in nodes:
        fairness[node] = 1
        try:
            goodness[node] = G.in_degree(node, weight='weight')*1.0/G.in_degree(node)
        except:
            goodness[node] = 0
    return fairness, goodness
```

```python
def compute_fairness_goodness(G):
    fairness,goodness=initiliaze_scores(G)
    
    nodes = G.nodes()
    iter = 0
    while iter < 100:
        df = 0
        dg = 0

        for node in nodes:
            inedges = G.in_edges(node, data='weight')   
            g = 0
            for edge in inedges:
                g += fairness[edge[0]]*edge[2]#["weight"]

            try:
                dg += abs(g/len(inedges) - goodness[node])
                goodness[node] = g/len(inedges)
            except:
                pass

        for node in nodes:
            outedges = G.out_edges(node, data='weight')
            f = 0
            for edge in outedges:
                f += 1.0 - abs(edge[2] - goodness[edge[1]])/2.0
            try:
                df += abs(f/len(outedges) - fairness[node])
                fairness[node] = f/len(outedges)
            except:
                pass
        
        if df < math.pow(10, -6) and dg < math.pow(10, -6):
            break
        iter+=1
    
    return fairness, goodness
```

```python
G = nx.DiGraph()

f = open("E:\\Academics\\VIT\\Project\\SIN\\Trust network\\Data\\soc-sign-bitcoinotc.csv","r")
for l in f:
    ls = l.strip().split(",")
    G.add_edge(ls[0], ls[1], weight = float(ls[2])/10)
f.close()
```

```python
fairness, goodness = compute_fairness_goodness(G)
```

```python
print('Fairness: ')
print(fairness['1'])
print('Goodness: ')
print(goodness['1'])
```

The fairness tells us whose opinion or rating to trust, whereas goodness tells us how well they are rated. That is, higher the fairness, more we can trust their opinion. Higher their goodness, more other people have rated them good.

"If a user consistently rates a 5 star-deserving product as 5 star and 1 star-deserving product as 1-star, then the user is fair. And a high quality (good) product gets high ratings from highly fair users"
